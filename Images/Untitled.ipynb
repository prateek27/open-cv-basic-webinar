```python
import cv2
import matplotlib.pyplot as plt

```

```python
# Camera Object
camera = cv2.VideoCapture(0)

# Click a Picture from the camera return 
success,pic = camera.read()

# CLose the Camera
camera.release()

if(success):
    pic = cv2.cvtColor(pic,4)
    plt.imshow(pic)
    plt.show()
```

```python
# cb.lk/facemodel
pic = cv2.imread('../face_photo.jpg')
pic = cv2.cvtColor(pic,4)

face_model = cv2.CascadeClassifier("../FaceCascade/haarcascade_frontalface_alt.xml")
```

```python
faces = face_model.detectMultiScale(pic)


for f in faces:
    face = f
    x,y,w,h = face
    cv2.rectangle(pic,(x,y),(x+w,y+h),(0,255,0),3)
    plt.imshow(pic)
    plt.axis("off")
plt.show()

```

```python
new_img = cv2.rectangle(pic,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(new_img)
plt.axis("off")
plt.show()
```
